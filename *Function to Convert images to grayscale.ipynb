```python
import os, glob, cv2
from PIL import Image
from os import mkdir

# these are the libraries important to the function
""" 
/**
*!  @brief Function to convert images to grayscale
**  @param path: Path to the images
**  @param save_path: Path to save the images
* @return void
*/ 

here, user needs to understand some variables that are important to the function:
lets understand that:
there are 2 variables for every image path and they are:
/**
*! 1. Input Path : 
** this is the path where our colored images are stored, we give this as input to the function, so that this converts the images to grayscale
*! 2. Output Path : 
** this is the path where we want to save the grayscale images, we give this as input to the function, so that this saves the images in the path that we gave as input
*/ 
"""
def convert_gray(path, save_path, save_name):
    try:
        #outputPath = input("Enter the path to save the images: ") # enter rough/Gray_2/
        mkdir(save_path)
    except:
        print ("Directory already exist, images will be written in same folder")
    
    i = 0
    for images in path:
        img = cv2.imread(images)
        #cv2.imshow('rough', img)
        print(img)
        gray_images = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        #cv2.imshow('Gray Images', gray_images)
        cv2.imwrite(save_path + save_name +"%02i.jpg" %i, gray_images)
        i += 1
        cv2.waitKey(500)
        cv2.destroyAllWindows()
    return 0
```

```python
path = glob.glob('rough/1/*.jpg') # this is the main path of the folder with images, not the path of subfolders of images
#save_path = 'rough/Gray_2/'
save_path = input("Enter the path to save the images: ")
save_name = input("Enter the name of the images: ")
#outputpath = rough/gray_2/
convert_gray(path, save_path, save_name)
```
